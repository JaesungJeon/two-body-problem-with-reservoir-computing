$\frac{dx}{dt} = a(y-z)$

$\frac{dy}{dt} = x(r-z)-y$

$\frac{dz}{dt} = xy-bz$


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
a,b,r = 10,8/3,28
x0,y0,z0 = 1,0,1
start = 0
end = 10
step = 1000
h = end/step
t = np.linspace(start, end,step)
X,Y,Z = [x0],[y0],[z0]
```

```python
for i in range(len(t)-1):
    x,y,z = X[i],Y[i],Z[i]
    nx = x+h* (a*(y-x))
    ny = y+h*(x*(r-z)-y)
    nz = z+h*(x*y-b*z)
    X.append(nx)
    Y.append(ny)
    Z.append(nz)

```

```python
plt.plot(X,Y,'b')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
plt.plot(X,Z,'b')
plt.xlabel('X')
plt.ylabel('Z')

plt.show()
```
