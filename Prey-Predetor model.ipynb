$\frac{dx}{dt} = ax-bxy$

$\frac{dy}{dt} = cxy-dy$

$x = $ population of prey
$\\ y = $ population of predetor

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#intial value
start = 0
end = 50
step = 1000
t = np.linspace(start,end,step)
h = end/step
#intial population
x0 =50 #prey
y0 = 10 #predetor
#coefficient
a = 0.4
b = 0.4

c = 0.09
d = 2

X,Y = [x0], [y0]

#Calculation
for i in range(len(t)-1):
    x,y = X[i],Y[i]
    nx = x + h*(a*x - b*x*y)
    ny = y + h*(c*x*y - d*y)

    X.append(nx)
    Y.append(ny)

plt.plot(t,X, label = 'prey')
plt.plot(t,Y, label = 'predetor')
plt.set(xlabel = 'TIme, t', ylabel = 'Population, P',title = 'Predetor vs. Prey')
plt.legend()
plt.show()
'''
plt.title('Predetor vs. prey')
plt.plot(t,X, label = 'prey')
plt.plot(t,Y, label = 'predetor')
plt.legend()
plt.show()
'''
```
